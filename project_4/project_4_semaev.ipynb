### Таблица calls

#### Проверка данных в столбце 'duration'

```python
# в столбце 'duration' стоит оценить количество звонков с продолжительностью 0 минут

len(calls[calls['duration'] == 0])
```

Довольно большое количество - порядка 25%. Возможно, это действительно безответные звонки. В наше время, когда многие предпочитают звонкам чаты, мессенджеры и голосовые сообщения внутри интернет-коммуникаций - не вижу удивительного в таком количестве нулевых значений. Но для проверки данных все же построим гистограмму по количеству пропущенных звонков для кажого пользователя.

```python
# на всякий случай посмотрим какие абоненты (по user_id) 
# и в каком количестве имеют нулевые звонки

zero_calls = calls[calls['duration'] == 0].groupby('user_id')['duration'].count()
zero_calls = zero_calls.reset_index()
```

```python
# построим диаграмму 

g=sns.relplot(data=zero_calls, x='user_id', y='duration', color='red')
sns.rugplot(data=zero_calls, x='user_id', y='duration', color='red')
g.fig.set_size_inches(12,8)
plt.title('Распределение безответных звонков по пользователям ', fontsize=14)
plt.xlabel('Пользователи', fontsize=12)
plt.ylabel('Количество безответных звонков', fontsize=12)
plt.show()
```

В целом - данные довольно равномерно распределены по пользователям и можно оставить их без изменений. 

Но мы поступим иначе - заменим 0 на Nan и удалим строки с этими данными: такие данные будут скорее выбросами. 

```python
calls['duration'] = calls['duration'].replace(0, np.NaN)
calls = calls.dropna()
```

#### Замена типа данных и округление времени звонков 

```python
# заменим тип данных в столбцах 'call_date' таблицы users на datetime.

calls['call_date'] = pd.to_datetime(calls['call_date'], format = '%Y-%m-%d')
```

```python
# округлим данные о времени звонков в большую сторону 
# и сразу же перведем данные в целочисленный тип

calls['duration'] = calls['duration'].apply(np.ceil).astype(int)
```

```python
calls.info()
```

#### Подсчет количества звонков по месяцам

```python
# добавим новый столбец с названием месяца 

calls['month'] = calls['call_date'].dt.month_name()
```

```python
# посчитаем количество звонков по месяцам 

calls_per_month = calls['month'].value_counts()
calls_per_month
```

Топ-3 месяца по количеству звонков - декабрь, ноябрь, октябрь. Предновогодние месяцы самые активные.

В то же время - после НГ происходит явный спад в желании общаться голосом. 

#### Подсчет количества минут по месяцам

```python
minutes_per_month = calls.groupby('month')['duration'].sum().sort_values(ascending=False)
minutes_per_month
```

Все в точности похоже на ситуйцию с количеством звонков - больше и дольше всего разговаривают по телефону в предновогодние месяцы. После НГ - спад и затем снова увеличается количество минут.

#### Барплот для количества звонков и минут

Для большей наглядности визуализируем данные вычислений

```python
# создаем общую таблицу с данными по количеству звонков и минут 

calls_with_minutes = pd.concat([minutes_per_month, calls_per_month], axis = 1)
calls_with_minutes = calls_with_minutes.rename(columns = {'duration': 'Количество минут', 'call_month': 'Количество звонков'})

# создаем сводный график по месяцам для количества звонков и минут

ax = calls_with_minutes.plot.barh(figsize=(12,8), color=['r','y'], width=0.9)

for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+700, i.get_y()+.18, \
            str(round((i.get_width()), 2)), fontsize=11, color='dimgrey')

plt.title('Количество минут и звонков по месяцам', fontsize=14)
plt.xlabel("Количество", fontsize=12)
plt.ylabel("Месяцы",fontsize=12)
plt.show()
```

Полная корреляция - чем больше, звонков - тем больше минут.

### Подсчет помесячной выручки с каждого пользователя

```python
# создадим сводную таблицу по звонкам для каждого user_id и месяца

calls_by_month_users = calls.pivot_table(index=['user_id', 'month'],  values=['duration'], aggfunc='sum').reset_index()
calls_by_month_users.head()
```

```python
# создадим сводную таблицу по сообщениям для каждого user_id и месяца

messages_by_month_users = messages.pivot_table(index=['user_id', 'month'],  values=['id'], aggfunc='count').reset_index()
messages_by_month_users.head()
```

```python
# создадим сводную таблицу по потраченным мб для каждого user_id и месяца

internet_by_month_users = internet.pivot_table(index=['user_id', 'month'], \
                                               values=['mb_used'], aggfunc='sum').reset_index()
internet_by_month_users.head()
```

```python
# создадим временную таблицу путем объединения таблиц с данными о количестве минут, смс и мб

temp = reduce(lambda left,right: pd.merge(left, right, \
                                             on=['user_id', 'month'], how='left'), \
                 [calls_by_month_users, messages_by_month_users, internet_by_month_users])

temp.head()
```

```python
temp = temp.rename(columns={'duration':'minutes', 'id':'messages'})
temp.head()
```

```python
# создадим временную таблицу путем объединения таблиц users и tariffs 

temp_2 = pd.merge(users, tariffs, on='tariff', how='left')
temp_2.head() 
```

```python
# объединяем две временные таблицы и получаем большой сводный датафрейм 

all_in_one = pd.merge(temp, temp_2, on='user_id', how='outer')
all_in_one.head()
```

Создадим ряд функций для рассчета превышений по тарифам

```python
# функция для подсчета минут сверх тарифа

def minutes_over_tariff(row):
    if row['tariff'] == 'smart':
        result = row['minutes'] - 500
        return np.where(result > 0, result, 0)
    else:
        result = row['minutes'] - 3000
        return np.where(result > 0, result, 0)
    

# функция для подсчета смс сверх тарифа   

def messages_over_tariff(row):
    if row['tariff'] == 'smart':
        result = row['messages'] - 50
        return np.where(result > 0, result, 0)
    else:
        result =  row['messages'] - 1000
        return np.where(result > 0, result, 0)
    

# функция для подсчета мб сверх тарифа   

def internet_over_tariff(row):
    if row['tariff'] == 'smart':
        result = row['mb_used'] - 15360
        return np.where(result > 0, result, 0)
    else:
        result = row['mb_used'] - 30720
        return np.where(result > 0, result, 0)
    
    
    
# функция для подсчета прибыли сверх абонентской платы 

def month_profit(row):
    sum_per_month = 0
    if row['tariff'] == 'smart':
        sum_per_month += np.ceil(row['minutes_over']) * 3
        sum_per_month += row['messages_over'] * 3
        sum_per_month += np.ceil(row['mb_over']/1024) * 200
        return sum_per_month
    
    else:
        sum_per_month += np.ceil(row['minutes_over']) * 1
        sum_per_month += row['messages_over'] * 1
        sum_per_month += np.ceil(row['mb_over']/1024) * 150
        return sum_per_month
```

```python
# создадим новые столбцы с данными по превышению 

all_in_one['minutes_over'] = all_in_one.apply(minutes_over_tariff, axis=1).astype(int)
all_in_one['messages_over'] = all_in_one.apply(messages_over_tariff, axis=1).astype(int)
all_in_one['mb_over'] = all_in_one.apply(internet_over_tariff, axis=1).astype(int)

# создаем столбец с суммой дохода сверх тарифа 
all_in_one['over_profit'] = all_in_one.apply(month_profit, axis=1)


# создадим столбец с общей суммой затрат 

all_in_one['total_profit'] = all_in_one['rub_monthly_fee'] + all_in_one['over_profit']
all_in_one.sample(5)
```

   - количество сообщений сверх тарифа

## Шаг 4. Проверка гипотез

### Гипотеза 1. Cредняя выручка пользователей тарифов «Ultra» и «Smart» различается

Нулевой гипотезой будет выражение "Средние наших выборок равны". 

Альтернативной гипотезой будет гипотеза задания -  "Средняя выручка пользователей тарифов «Ultra» и «Smart» различается"

```python
# уровень значимости - 0.05 
alpha = 0.05


results = st.ttest_ind(smart_only['total_profit'], ultra_only['total_profit'], equal_var = False)
print('P-value:', results.pvalue)

if (results.pvalue < alpha):
    print('Нулевая гипотеза отвергнута')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Значит сформулированная альтернативная гипотеза не отвергается - средние наших выборок отличаются. 

### Гипотеза 2. Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Нулевой гипотезой зададим выражение "Средняя выручка пользователей из Москвы равна выручке пользователей из других регионов".

Альтернативной гипотезой будет гипотеза задания - "Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов". 

```python
# создадим две таблицы:
# moscow_only - только клиенты из Москвы
# other_only - только клиенты не относящиеся к Москве

moscow_only = all_in_one[all_in_one['city'] == 'Москва']
other_only = all_in_one[all_in_one['city'] != 'Москва']

display(moscow_only.sample(5))
display(other_only.sample(5))
```

```python
# проверим соотношения пользователей разных тарифов в кождой из выборок

display('Соотношение клиентов разных тарифов в Москве', moscow_only.pivot_table(index = 'tariff', values = 'total_profit', aggfunc = 'count'))
display('Соотношение клиентов разных тарифов в других регионах', other_only.pivot_table(index = 'tariff', values = 'total_profit', aggfunc = 'count'))
```

Судя по данным - распределение клиентов по тарифам не зависит критично от региона. В Москве, что не может быть неожиданностью, чуть больше людей предпочитают дорогой тариф. 

```python
# уровень значимости - 0.05 
alpha = 0.05

results = st.ttest_ind(moscow_only['total_profit'], other_only['total_profit'], equal_var = False)
print('P-value:', results.pvalue)

if (results.pvalue < alpha):
    print('Нулевая гипотеза отвергнута')
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Полученное p-value =  0.43, а значит есть вероятность того, что отличие средних значений выборок случайно и нельзя отвергнуть нулевую гипотезу, которая говорит о равенстве средних значений выручки в Москве и регионах.

Альтернативная гипотеза отвергнута.
