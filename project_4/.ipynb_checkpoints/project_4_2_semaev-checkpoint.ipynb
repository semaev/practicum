```python
# –ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–≤–æ–Ω–∫–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º 

calls_per_month = calls['month'].value_counts()
calls_per_month
```

```python
# —Å–æ–∑–¥–∞–µ–º –æ–±—â—É—é —Ç–∞–±–ª–∏—Ü—É —Å –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∑–≤–æ–Ω–∫–æ–≤ –∏ –º–∏–Ω—É—Ç 

calls_with_minutes = pd.concat([minutes_per_month, calls_per_month], axis = 1)
# —Å–æ–∑–¥–∞–µ–º —Å–≤–æ–¥–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –ø–æ –º–µ—Å—è—Ü–∞–º –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–≤–æ–Ω–∫–æ–≤ –∏ –º–∏–Ω—É—Ç

ax = calls_with_minutes.plot.bar(figsize=(14,8), color=['r','y'], width=0.9, rot=30)
ax.legend(['–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç', '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–≤–æ–Ω–∫–æ–≤']);

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., \
                                      p.get_height()), ha = 'center', va = 'center', \
                xytext = (0, 10), textcoords = 'offset points')

plt.title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç –∏ –∑–≤–æ–Ω–∫–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º', fontsize=14)
plt.xlabel('–ú–µ—Å—è—Ü—ã', fontsize=12)
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ',fontsize=12)
plt.show()
```

–î–æ–≤–æ–ª—å–Ω–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ –∏ –±–µ–∑ –∫–∞–∫–∏—Ö –ª–∏–±–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π. 

–ò–∑–±–∞–≤–∏–º—Å—è –æ—Ç –Ω—É–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π.

```python
internet['mb_used'] = internet['mb_used'].replace(0, np.NaN)
internet = internet.dropna()
```

#### –ó–∞–º–µ–Ω–∞ —Ç–∏–ø–∞ –¥–∞–Ω–Ω—ã—Ö, –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω—ã—Ö –º–±

```python
# –ø–µ—Ä–µ–≤–µ–¥–µ–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–∞ 'session_date' –≤  datetime

internet['session_date'] = pd.to_datetime(internet['session_date'], format = '%Y-%m-%d')
```

```python
# –æ–∫—Ä—É–≥–ª–∏–º –¥–∞–Ω–Ω—ã–µ –æ –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω—ã—Ö –º–± –≤ –±–æ–ª—å—à—É—é —Å—Ç–æ—Ä–æ–Ω—É 
# –∏ —Å—Ä–∞–∑—É –∂–µ –ø–µ—Ä–≤–µ–¥–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã–π —Ç–∏–ø

internet['mb_used'] = internet['mb_used'].apply(np.ceil).astype(int)
```

#### –ü–æ–¥—Å—á–µ—Ç –æ–±—ä–µ–º–∞ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫–∞ –ø–æ –º–µ—Å—è—Ü–∞–º

```python
# –¥–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –Ω–∞–∑–≤–∞–Ω–∏–µ–º –º–µ—Å—è—Ü–∞ 

internet['month'] = internet['session_date'].dt.month_name()
```

```python
# –ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω—ã—Ö –º–± –ø–æ –º–µ—Å—è—Ü–∞–º 

mb_per_month = internet.groupby('month')['mb_used'].sum().sort_values(ascending=False)
mb_per_month
```

–î–æ–≤–æ–ª—å–Ω–æ —è–≤–Ω–æ–µ —É–≤–µ–ª–∏—á–µ–Ω–∏–µ —Ç—Ä–∞—Ñ–∏–∫–∞ —Å —è–Ω–≤–∞—Ä—è –ø–æ –¥–µ–∫–∞–±—Ä—å. –ò —Å–Ω–æ–≤–∞ - —Å–∞–º—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã–µ –º–µ—Å—è—Ü—ã: –æ–∫—Ç—è–±—Ä—å, –Ω–æ—è–±—Ä—å –∏ –¥–µ–∫–∞–±—Ä—å.

### –¢–∞–±–ª–∏—Ü–∞ messages

#### –ó–∞–º–µ–Ω–∞ —Ç–∏–ø–∞ –¥–∞–Ω–Ω—ã—Ö

```python
# –ø–µ—Ä–µ–≤–µ–¥–µ–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–∞ 'message_date' –≤  datetime

messages['message_date'] = pd.to_datetime(messages['message_date'], format = '%Y-%m-%d')
```

```python
messages.info()
```

#### –ü–æ–¥—Å—á–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π –ø–æ –º–µ—Å—è—Ü–∞–º

```python
# –¥–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –Ω–∞–∑–≤–∞–Ω–∏–µ–º –º–µ—Å—è—Ü–∞ 

messages['month'] = messages['message_date'].dt.month_name()
```

```python
# –ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–º—Å –ø–æ –º–µ—Å—è—Ü–∞–º 

messages_per_month = messages['month'].value_counts()
messages_per_month
```

–í—Å–µ –≤ —Ç–æ—á–Ω–æ—Å—Ç–∏, –∫–∞–∫ –∏ —Å –æ—Å—Ç–∞–ª—å–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ - —É–≤–µ–ª–∏—á–µ–Ω–∏–µ –æ—Ç —è–Ω–≤–∞—Ä—è –∫ –¥–µ–∫–∞–±—Ä—é.

### –¢–∞–±–ª–∏—Ü–∞ tariffs

```python
# –ø–µ—Ä–µ–∏–º–µ–Ω—É–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ —Ç–∞—Ä–∏—Ñ–æ–≤
tariffs = tariffs.rename(columns = {'tariff_name': 'tariff'})
```

### –ü–æ–¥—Å—á–µ—Ç –ø–æ–º–µ—Å—è—á–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ —Å –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø–æ –∑–≤–æ–Ω–∫–∞–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ user_id –∏ –º–µ—Å—è—Ü–∞

calls_by_month_users = calls.pivot_table(index=['user_id', 'month'],  values=['duration'], aggfunc='sum').reset_index()
calls_by_month_users.head()
```

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø–æ —Å–æ–æ–±—â–µ–Ω–∏—è–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ user_id –∏ –º–µ—Å—è—Ü–∞

messages_by_month_users = messages.pivot_table(index=['user_id', 'month'],  values=['id'], aggfunc='count').reset_index()
messages_by_month_users.head()
```

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø–æ –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω—ã–º –º–± –¥–ª—è –∫–∞–∂–¥–æ–≥–æ user_id –∏ –º–µ—Å—è—Ü–∞

internet_by_month_users = internet.pivot_table(index=['user_id', 'month'], \
                                               values=['mb_used'], aggfunc='sum').reset_index()
internet_by_month_users.head()
```

```python
# —Å–æ–∑–¥–∞–¥–∏–º –≤—Ä–µ–º–µ–Ω–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø—É—Ç–µ–º –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è —Ç–∞–±–ª–∏—Ü —Å –¥–∞–Ω–Ω—ã–º–∏ –æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –º–∏–Ω—É—Ç, —Å–º—Å –∏ –º–±

temp = reduce(lambda left,right: pd.merge(left, right, \
                                             on=['user_id', 'month'], how='outer'), \
                 [calls_by_month_users, messages_by_month_users, internet_by_month_users])

temp.head()
```

```python
# –æ–±—ä–µ–¥–∏–Ω—è–µ–º –¥–≤–µ –≤—Ä–µ–º–µ–Ω–Ω—ã–µ —Ç–∞–±–ª–∏—Ü—ã –∏ –ø–æ–ª—É—á–∞–µ–º –±–æ–ª—å—à–æ–π —Å–≤–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º 

all_in_one = pd.merge(temp, temp_2, on='user_id', how='outer')
all_in_one.head()
```

–°–æ–∑–¥–∞–¥–∏–º —Ä—è–¥ —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è —Ä–∞—Å—Å—á–µ—Ç–∞ –ø—Ä–µ–≤—ã—à–µ–Ω–∏–π –ø–æ —Ç–∞—Ä–∏—Ñ–∞–º

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –º–∏–Ω—É—Ç —Å–≤–µ—Ä—Ö —Ç–∞—Ä–∏—Ñ–∞

def minutes_over_tariff(row):
    if row['tariff'] == 'smart':
        result = row['minutes'] - 500
        return np.where(result > 0, result, 0)
    else:
        result = row['minutes'] - 3000
        return np.where(result > 0, result, 0)
    

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ —Å–º—Å —Å–≤–µ—Ä—Ö —Ç–∞—Ä–∏—Ñ–∞   

def messages_over_tariff(row):
    if row['tariff'] == 'smart':
        result = row['messages'] - 50
        return np.where(result > 0, result, 0)
    else:
        result =  row['messages'] - 1000
        return np.where(result > 0, result, 0)
    

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –º–± —Å–≤–µ—Ä—Ö —Ç–∞—Ä–∏—Ñ–∞   

def internet_over_tariff(row):
    if row['tariff'] == 'smart':
        result = row['mb_used'] - 15360
        return np.where(result > 0, result, 0)
    else:
        result = row['mb_used'] - 30720
        return np.where(result > 0, result, 0)
    
    
    
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –ø—Ä–∏–±—ã–ª–∏ —Å–≤–µ—Ä—Ö –∞–±–æ–Ω–µ–Ω—Ç—Å–∫–æ–π –ø–ª–∞—Ç—ã 

def month_profit(row):
    sum_per_month = 0
    if row['tariff'] == 'smart':
        sum_per_month += np.ceil(row['minutes_over']) * 3
        sum_per_month += row['messages_over'] * 3
        sum_per_month += np.ceil(row['mb_over']/1024) * 200
        return sum_per_month
    
    else:
        sum_per_month += np.ceil(row['minutes_over']) * 1
        sum_per_month += row['messages_over'] * 1
        sum_per_month += np.ceil(row['mb_over']/1024) * 150
        return sum_per_month
```

<div style="border:solid orange 2px; padding: 20px"> <b>–ß—Ç–æ –º–æ–∂–Ω–æ —É–ª—É—á—à–∏—Ç—å</b><br>
–¢–∞–∫ –∑–¥–æ—Ä–æ–≤–æ —Å–æ–±—Ä–∞–ª –≤—Å–µ –¥–∞–Ω–Ω—ã–µ –≤ –æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ, —á—Ç–æ–±—ã —Ä—É—á–∫–∞–º–∏ –ø–æ–¥—Å—Ç–∞–≤–ª—è—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è ?! –ó–∞—á–µ–º –µ—Å–ª–∏, –æ–Ω–∏ –∑–∞–ø–∏—Å–∞–Ω—ã –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö, –≤—ã–ø–æ–ª–Ω—è–π –æ–ø–µ—Ä–∞—Ü–∏–∏ –Ω–∞–¥ —Å—Ç–æ–ª–±—Ü–∞–º–∏ –∏ –Ω–∏—á–µ–≥–æ –Ω–µ –Ω–∞–¥–æ –ø–æ–¥—Å—Ç–∞–≤–ª—è—Ç—å (–ø—Ä–∏–º–µ—Ä –Ω–∏–∂–µ)</div> <br>

```python
all_in_one\
    .assign(sms_payment = lambda row: np.where(row['messages'] - \
                                               row['messages_included'] > 0, \
                                               row['messages'] - row['messages_included'], 0) * \
            row['rub_per_message'])
```

### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞


–ë–ª–∞–≥–æ–¥–∞—Ä—é. –û–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑—É—é—Å—å —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–µ–π. –í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ - —Å–æ–≥–ª–∞—Å–µ–Ω, –º–æ–∂–Ω–æ –±—ã–ª–æ –∏ —Å–æ–∫—Ä–∞—Ç–∏—Ç—å. 


</div>

```python
# —Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã —Å –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –ø—Ä–µ–≤—ã—à–µ–Ω–∏—é 

all_in_one['minutes_over'] = all_in_one.apply(minutes_over_tariff, axis=1).astype(int)
all_in_one['messages_over'] = all_in_one.apply(messages_over_tariff, axis=1).astype(int)
all_in_one['mb_over'] = all_in_one.apply(internet_over_tariff, axis=1).astype(int)

# —Å–æ–∑–¥–∞–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å —Å—É–º–º–æ–π –¥–æ—Ö–æ–¥–∞ —Å–≤–µ—Ä—Ö —Ç–∞—Ä–∏—Ñ–∞ 
all_in_one['over_profit'] = all_in_one.apply(month_profit, axis=1)


# —Å–æ–∑–¥–∞–¥–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –æ–±—â–µ–π —Å—É–º–º–æ–π –∑–∞—Ç—Ä–∞—Ç 

all_in_one['total_profit'] = all_in_one['rub_monthly_fee'] + all_in_one['over_profit']
all_in_one.sample(5)
```
