#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Обучение

```python
data = data.drop(['std'], axis=1)
```

```python
# создадим функцию, чтобы прибавить к таблице четыре новых календарных признака: 
# месяц, день, день недели и час

def make_features(data, max_lag, rolling_mean_size):
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    data.dropna(inplace=True)
```

<div class="alert alert-block alert-info">
<b>Совет: </b> Желательно было написать функцию таким образом, чтобы она возвращала новый датафрейм, а не изменяла глобальную переменную.
</div>

```python
# прогоним наши данные через функцию make_features()

make_features(data, 10, 10)
```

```python
# разделим данные на обучающую т тестовую выборки

train, test = train_test_split(data, shuffle=False, test_size=0.1)
train = train.dropna()
test = test.dropna()

# проверим размеры выборок 
display(train.shape, test.shape)
```

```python
# выделим в наших выборках признаки и целевой признак

X_train, X_test = train.drop(['num_orders'], axis=1), test.drop(['num_orders'], axis=1)
y_train, y_test = train['num_orders'],  test['num_orders']
```

<b>Ошибка:</b> Ты оставил в данных признак std, который посчитан на основе значения num_orders без сдвига, это утечка данных.
</div>

## Тестирование

Прогоним три модели без подбора параметров по нашим выборкам. 

```python
def train_and_show_rmse(model, title):
    model = model
    model.fit(X_train, y_train)
    train_prediction = model.predict(X_train)
    test_prediction = model.predict(X_test)
    print(title)
    print("RMSE обучающей выборки:", mean_squared_error(y_train, train_prediction)**0.5)
    print("RMSE тестовой выборки: ", mean_squared_error(y_test, test_prediction)**0.5)
```

```python
%%time
train_and_show_rmse(LinearRegression(), 'LinearRegression')
```

```python
%%time
train_and_show_rmse(RandomForestRegressor(), 'RandomForestRegressor')
```

```python
%%time
train_and_show_rmse(LGBMRegressor(), 'LGBMRegressor')
```

А теперь подберем параметры для каждой модели, обучим и посчитаем RMSE.

```python
tscv = TimeSeriesSplit(n_splits=5)
```

```python
lr_params = {
    'fit_intercept':[True,False], 
    'normalize':[True,False], 
    'copy_X':[True, False]
             }
```

```python
%%time
lr_grid = GridSearchCV(LinearRegression(), param_grid = lr_params, \
                       refit=False, scoring='neg_mean_squared_error', cv=tscv)
lr_grid.fit(X_train, y_train)

lr_best_params = lr_grid.best_params_
lr_m = LinearRegression(**lr_best_params)
lr_m.fit(X_train, y_train)
    
lr_test_pred = lr_m.predict(X_test)

print('Лучшие гиперпараметры: ', lr_best_params)
print("RMSE тестовой выборки: ", mean_squared_error(y_test, lr_test_pred)**0.5)
```

<div class="alert alert-block alert-success">
<b>Успех:</b> Молодец, что для подбора параметров использовал метод основанный на кросс-валидации с корректной метрикой и подходящим под данные сплиттером.
</div>

```python
forest_params = {
    'n_estimators' : range(50, 101, 10),
    'max_depth' : range(5, 15, 2), 
    'verbose' : [0], 
    'random_state' : [1511]
                 }
```

```python
%%time
forest_grid = GridSearchCV(RandomForestRegressor(), param_grid = forest_params, \
                           refit=False, scoring='neg_mean_squared_error', cv=tscv)
forest_grid.fit(X_train, y_train)

forest_best_params = forest_grid.best_params_
forest_m = RandomForestRegressor(**forest_best_params)
forest_m.fit(X_train, y_train)
    
forest_test_pred = forest_m.predict(X_test)

print('Лучшие гиперпараметры: ', forest_best_params)
print("RMSE тестовой выборки: ", mean_squared_error(y_test, forest_test_pred)**0.5)
```

```python
lgb_params = {
    'learning_rate': np.logspace(-3, 0, 5), 
    'n_estimators': range(20, 61, 20), 
    'num_leaves': range(20, 41, 10)
              }
```

```python
%%time
lgb_grid = GridSearchCV(LGBMRegressor(), param_grid = lgb_params, \
                        refit=False, scoring='neg_mean_squared_error', cv=tscv)
lgb_grid.fit(X_train, y_train)

lgb_best_params = lgb_grid.best_params_
lgb_m = LGBMRegressor(**lgb_best_params)
lgb_m.fit(X_train, y_train)
    
lgb_test_pred = lgb_m.predict(X_test)

print('Лучшие гиперпараметры: ', lgb_best_params)
print("RMSE тестовой выборки: ", mean_squared_error(y_test, lgb_test_pred)**0.5)
```
