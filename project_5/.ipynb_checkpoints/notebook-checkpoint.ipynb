## Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

### Работа с пропусками

#### Пропуски в столбце 'name'

```python
# выведем обе строки, где есть пропуски в столбце 'name'

games[games['name'].isna()]
```

Судя по всему - придется избавиться от этих строк: нет информации не только по названию игры, но и по остальным важным параметрам. 

```python
# перезапишем датасет без этих строк 
games = games[games['name'].notna()]

# проверяем 
games['name'].isna().sum()
```

```python
# так же мы сразу избавились от 2-х пропусков в столбце 'genre'

games['genre'].isna().sum()
```

#### Пропуски в 'year_of_release'

```python
# изучим строки, где пропущен год выпуска игры 

games[games['year_of_release'].isna()].info()
```

Рассмотрим распределение пропусков данного столбца по категориальными данным - жанрам и платформам.

```python
# сгруппируем даные с пропущенными годами по жанрам 
genre_non_years = games[games['year_of_release'].isna()].groupby('genre')['name'] \
.count().sort_values(ascending=False).reset_index()

# построим график распределения данных по жанрам 
plt.figure(figsize=(12, 8))
g=sns.barplot(data=genre_non_years, x='genre', y='name', color='tomato')

for p in g.patches:
    g.annotate(str(int(p.get_height())), \
               (p.get_x() + p.get_width() / 2.,\
                p.get_height()), ha='center', va='center', xytext=(0, 10), \
               textcoords='offset points')

plt.title('Распределение пропущенных данных в столбце year_of_release по жанрам', fontsize=14)
plt.xlabel('Жанры игр', fontsize=12)
plt.ylabel('Количество строк с пропусками', fontsize=12)
plt.show()
```

Никакой зависимости от жанра игры с пропусками в столбце 'year_of_release' не обнаруживается. Пропуски есть во всех жанрах.

Значит нужно попробовать другие варианты. Рассмотрим второй столбец с категориальными данными - платформы.

```python
# сгруппируем даные с пропущенными годами по платформам
platform_non_years = games[games['year_of_release'].isna()].groupby('platform')['name'] \
.count().sort_values(ascending=False).reset_index()

# построим график распределения данных по платформам 
plt.figure(figsize=(12, 8))
g=sns.barplot(data=platform_non_years , x='platform', y='name', color='darkorange')

for p in g.patches:
    g.annotate(str(int(p.get_height())), \
               (p.get_x() + p.get_width() / 2.,\
                p.get_height()), ha='center', va='center', xytext=(0, 10), \
               textcoords='offset points')


plt.title('Распределение пропущенных данных в столбце year_of_release по платформам', fontsize=14)
plt.xlabel('Платформы', fontsize=12)
plt.ylabel('Количество строк с пропусками', fontsize=12)
plt.show()
```

Данные пропущены примерно у половины платформ, но нет никаких заметных взаимосвязей с пропущенными данными. 

Скорее всего - эти проуски случайны. 

Получается выход один - удалить данные пропуски. К тому же - их всего 1.61% от общего числа данных в столбце.

```python
# сбрасывем 
games = games.dropna(subset=['year_of_release'])

# проверяем
games['year_of_release'].isna().sum()
```

Заметно, что пропуски в 'user_rating' так же имеют пересечения с пропусками в других рейтинговых столбцах: 'critic_score' и 'rating'.

Построим графики распределения пропусков в этих столбцах по жанрам и платформам. 

```python
# сгруппируем даные с пропущенными оценками пользователей по жанрам 
genre_non_user_score = games[games['user_score'].isna()].groupby('genre')['name'] \
.count().sort_values(ascending=False).reset_index()

# сгруппируем даные с пропущенными оценками критиков по жанрам 
genre_non_critic_score = games[games['critic_score'].isna()].groupby('genre')['name'] \
.count().sort_values(ascending=False).reset_index()

# сгруппируем даные с пропущенными рейтингами от организации ESRB по жанрам 
genre_non_rating = games[games['rating'].isna()].groupby('genre')['name'] \
.count().sort_values(ascending=False).reset_index()

# сведем все таблицы в одну
non_by_genre = pd.concat([genre_non_user_score.set_index('genre'), \
                          genre_non_critic_score.set_index('genre'), \
                          genre_non_rating.set_index('genre')], axis=1)

# построим график распределения пропущеных данных по жанрам 
ax = non_by_genre.plot.bar(figsize=(14,8), color=['orangered','darkorchid','aqua'], width= 0.9)
ax.legend(['Пропуски в user_score', 'Пропуски в critic_score', 'Пропуски в rating']);

# подписываем значения 
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), \
               (p.get_x() + p.get_width() / 2.,\
                p.get_height()), ha='center', va='center', xytext=(0, 10), \
               textcoords='offset points', rotation=30)

# подписываем оси и задаем доп параметры
plt.title('Распределение пропущенных данных по жанрам', fontsize=14)
plt.xlabel('Жанры игр', fontsize=12)
plt.xticks(rotation=0)
plt.ylabel('Количество строк с пропусками', fontsize=12)
plt.show()
```

Прослеживается очень четкая взаимосвяь между попусками в 'user_score' и 'critic_score'. Попусков в 'rating' немного меньше, но они так же распределены похожим способом, как и у двух других исследуемых столбцов.

### Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков

Для рассмотрения возьмем PS4 и построим диаграмму рассеяния. 

```python
# возьмем выборку для PS4 без пропусков в столбцах 'user_score' и 'critic_score'

ps4_only = games[(games['user_score'].notna()) & (games['critic_score'].notna())] \
.query('platform == "PS4"')
```

```python
# построим диаграмму рассеяния для пары "продажи - рейтинг критиков"
plt.figure(figsize=(14, 8))
g = sns.scatterplot(x='critic_score', y='total_sales', size='total_sales', 
                sizes=(50, 800), data=ps4_only, color='r', alpha=.8)

# подписываем оси и задаем доп параметры
plt.title('Распределение продаж по рейтингу критиков для PS4', fontsize=16)
plt.xlabel('Рейтинг критиков', fontsize=14)
plt.ylabel('Мировые продажи в млн.',fontsize=14)
plt.show()
```

По диаграммам распределения можно сделать вывод: 
- чем выше рейтинг критиков, тем больше продаж на платформе PS4 
- та же самая зависимость, но менее явная, прослеживается с рейтингами пользователей

```python
# построим матрицу корреляции мировых и региональных продаж с пользовательским рейтингом и рейтингом критиков

corr_matrix = games.query('platform == "PS4"')[['total_sales', 'na_sales', 'eu_sales', \
                                               'jp_sales', 'other_sales', 'critic_score', \
                                               'user_score']].corr()
fig, ax = plt.subplots(figsize=(14,10)) 
sns.heatmap(corr_matrix, annot=True, 
            cbar_kws={'orientation': 'horizontal', 
                      'label': 'Величина коэффициента корреляции Пирсона'},
            cmap='BuPu')
plt.title('Корреляция продаж с рейтингами критиков и пользователей для PS4', fontsize=16)
plt.show() 
```

Корреляции продаж по всем регионам игр на PS4:
- продажи - рейтинг критиков - положительная слабая корреляция (от 0.32 до 0.42)
- продажи - рейтинг пользователей - очень слабая (-0.07 до 0.14)

В данном случае можно сказать, что рейтинги критиков и пользователей практически не влияют на продажи игр для PS4 по всему миру.

Для оценки влияния рейтингов пользователей и критиков в целом на платформы - сделаем выборку необходимых данных из исходной таблицы: продажи общие, продажи по регионам, рейтинги и все это возьмем без пропусков в столбцах с рейтингами. 

И построим матрицу корреляции.

```python
games[['total_sales', 'critic_score', 'user_score']].corr()
# построим матрицу корреляции мировых и региональных продаж с пользовательским рейтингом и рейтингом критиков

corr_matrix = games[(games['user_score'].notna()) & (games['critic_score'].notna())] \
[['total_sales', 'na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'critic_score', \
  'user_score']].corr()
fig, ax = plt.subplots(figsize=(14,10)) 
sns.heatmap(corr_matrix, annot=True, 
            cbar_kws={'orientation': 'horizontal', 
                      'label': 'Величина коэффициента корреляции Пирсона'},
            cmap='YlOrBr')
plt.title('Корреляция продаж с рейтингами критиков и пользователей для всех платформ', fontsize=16)
plt.show() 
```

Как и в случае с частным примером (PS4) мы имеем картину:

## Шаг 5. Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нулевой гипотезой будет выражение: 
- "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые".

Альтернативной гипотезой будет гипотеза: 
- "Средние пользовательские рейтинги платформ Xbox One и PC различны"

```python
# уровень значимости 
alpha = 0.05


results = st.ttest_ind(games.query('platform == "XOne" and user_score != 0')['user_score'], \
                       games.query('platform == "PC" and user_score != 0')['user_score'], \
                       equal_var=False)
print('значение p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print('Нулевая гипотеза отвергнута')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Вполне предсказуемый результат - средние пользовательские рейтинги платформ Xbox One и PC отличаются.

### Средние пользовательские рейтинги жанров Action и Sports разные

Нулевой гипотезой будет выражение: 
- "Средние пользовательские рейтинги жанров Action и Sports одинаковые".

Альтернативной гипотезой будет гипотеза: 
- "Средние пользовательские рейтинги жанров Action и Sports различны"

```python
# уровень значимости
alpha = 0.05


results = st.ttest_ind(games.query('genre == "Action" and user_score != 0')['user_score'], \
                       games.query('genre == "Sports" and user_score != 0')['user_score'], \
                       nan_policy='omit')

print('Значение p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print('Нулевая гипотеза отвергнута')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Нам не удалось отпросить нудевую гипотезу и средние пользовательские рейтинги жанров Action и Sports вполне могут быть одинаковыми.

>**<span style="color:green">С формулировкой и проверкой гипотез у тебя полный порядок. Молодец, что ты учёл разницу дисперсий наших выборок и учел это при проведении теста, передав в параметр `equal_var` соответствующее значение. Обычно принято считать дисперсии отличными друг от друга, если они отличаются более чем на 5%, но в статистике есть специальные критерии (Левене, Бартлетта), которые позволяют проверить гипотезы о однородности дисперсий двух выборок. С этими критериями и их реализацией в пакете `stats` ты можешь ознакомиться самостоятельно. Отдельный респект за знание и использование параметра `nan_policy` — это очень удобно.</span>**
