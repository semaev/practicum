## Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

#### Пропуски в 'year_of_release'

```python
# изучим строки, где пропущен год выпуска игры 

games[games['year_of_release'].isna()].info()
```

Рассмотрим распределение пропусков данного столбца по категориальными данным - жанрам и платформам.

Никакой зависимости от жанра игры с пропусками в столбце 'year_of_release' не обнаруживается. Пропуски есть во всех жанрах.

```python
# сгруппируем даные с пропущенными годами по платформам
platform_non_years = games[games['year_of_release'].isna()].groupby('platform')['name'].count().sort_values(ascending=False).reset_index()

# построим график распределения данных по платформам 
plt.figure(figsize=(12, 8))
g=sns.barplot(data=platform_non_years , x='platform', y='name', color='darkorange')

for p in g.patches:
    g.annotate(str(int(p.get_height())), \
               (p.get_x() + p.get_width() / 2.,\
                p.get_height()), ha='center', va='center', xytext=(0, 10), \
               textcoords='offset points')


plt.title('Распределение пропущенных данных в столбце year_of_release по платформам' , fontsize=14)
plt.xlabel('Платформы', fontsize=12)
plt.ylabel('Количество строк с пропусками', fontsize=12)
plt.show()
```

Данные пропущены примерно у половины платформ, но нет никаких заметных взаимосвязей с пропущенными данными. 

Скорее всего - эти проуски случайны.

```python
# сбрасывем 
games = games.dropna(subset=['year_of_release'])

# проверяем
games['year_of_release'].isna().sum()
```

#### Пропуски в столбце 'user_score', 'critic_score' и  'rating'

Для начала разберемся со значениями tbd в столбце 'user_score'

```python
# посчитаем количество строк с tbd

len(games[games['user_score'] == 'tbd'])
```

Вот так новости. Еще 2376 пропусков прибыло. 

Заменяем tbd на Nan.

```python
games['user_score'] = games['user_score'].mask(games['user_score'] == 'tbd', np.nan) 
```

```python
# изучим строки, где пропущена оценка пользователей 

games[games['user_score'].isna()].info()
```

Заметно, что пропуски в 'user_rating' так же имеют пересечения с пропусками в других рейтинговых столбцах: 'critic_score' и 'rating'.

Построим графики распределения пропусков в этих столбцах по жанрам и платформам. 

```python
# сгруппируем даные с пропущенными оценками пользователей по жанрам 
genre_non_user_score = games[games['user_score'].isna()].groupby('genre')['name'] \
.count().sort_values(ascending=False).reset_index()

# сгруппируем даные с пропущенными оценками критиков по жанрам 
genre_non_critic_score = games[games['critic_score'].isna()].groupby('genre')['name'] \
.count().sort_values(ascending=False).reset_index()

# сгруппируем даные с пропущенными рейтингами от организации ESRB по жанрам 
genre_non_rating = games[games['rating'].isna()].groupby('genre')['name'] \
.count().sort_values(ascending=False).reset_index()

# сведем все таблицы в одну
non_by_genre = pd.concat([genre_non_user_score.set_index('genre'), \
                          genre_non_critic_score.set_index('genre'), \
                          genre_non_rating.set_index('genre')], axis=1)

# построим график распределения пропущеных данных по жанрам 
ax = non_by_genre.plot.bar(figsize=(14,8), color=['orangered','darkorchid','aqua'], width= 0.9)
ax.legend(['Пропуски в user_score', 'Пропуски в critic_score', 'Пропуски в rating']);

# подписываем значения 
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), \
               (p.get_x() + p.get_width() / 2.,\
                p.get_height()), ha='center', va='center', xytext=(0, 10), \
               textcoords='offset points', rotation=30)

# подписываем оси и задаем доп параметры
plt.title('Распределение пропущенных данных по жанрам', fontsize=14)
plt.xlabel('Жанры игр', fontsize=12)
plt.xticks(rotation=0)
plt.ylabel('Количество строк с пропусками', fontsize=12)
plt.show()
```

Прослеживается очень четкая взаимосвяь между попусками в 'user_score' и 'critic_score'. Попусков в 'rating' немного меньше, но они так же распределены похожим способом, как и у двух других исследуемых столбцов. 

Но вот взаимосвязи с каким либо определенным жанром не прослеживается. Хотя есть группа из 5 жанров (Action, Misc, Adventure, Sports и Role-Playing), где количество пропусков в 2 и больше раз превосходит количество пропусков в остальных 7 жанрах. 

Теперь построим график распредления этих пропусков по платформам.

```python
# сгруппируем даные с пропущенными оценками пользователей по платформам
genre_non_user_score = games[games['user_score'].isna()].groupby('platform')['name'] \
.count().sort_values(ascending=False).reset_index()

# сгруппируем даные с пропущенными оценками критиков по платформам
genre_non_critic_score = games[games['critic_score'].isna()].groupby('platform')['name'] \
.count().sort_values(ascending=False).reset_index()

# сгруппируем даные с пропущенными рейтингами от организации ESRB по платформам
genre_non_rating = games[games['rating'].isna()].groupby('platform')['name'] \
.count().sort_values(ascending=False).reset_index()

non_by_genre = pd.concat([genre_non_user_score.set_index('platform'), \
                          genre_non_critic_score.set_index('platform'), \
                          genre_non_rating.set_index('platform')], axis=1)

# построим график распределения пропущеных данных по платформам
ax = non_by_genre.plot.bar(figsize=(14,8), color=['orange','limegreen','darkviolet'], width= 0.9)
ax.legend(['Пропуски в user_score', 'Пропуски в critic_score', 'Пропуски в rating']);

# подписываем оси и задаем доп параметры
plt.title('Распределение пропущенных данных по платформам', fontsize=14)
plt.xlabel('Жанры игр', fontsize=12)
plt.xticks(rotation=30)
plt.ylabel('Количество строк с пропусками', fontsize=12)
plt.show()
```

Как и в случае с жанрами - нет какой-то одно платформы с ярко выраженными пропусками.

В данном случае выделяются 4 платформы с самым большим числом незаполненных данных - DS, PS, PS2 и PSP. 

Так же пропуски в 'user_score' и 'crtic_score' почти hавнозначны по платформам. 

Пропусков в 'rating' меньше (по сравнению с 'user_score' и 'crtic_score') по нескольким платформам: DS, Wii, PS2 и X360.


Из всего этого богатого анализа пропусков и с учетом большого процента пропусков (от 40% и выше) можно сделать только один вывод - не стоит их трогать. Заполнить их медианными значениями или 0 - значит сильно исказить. И избавиться от них тоже нельзя - будут потеряны большие массивы данных.

## Шаг 5. Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нулевой гипотезой будет выражение: 
- "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые".

Альтернативной гипотезой будет гипотеза: 
- "Средние пользовательские рейтинги платформ Xbox One и PC различны"

```python
# уровень значимости 
alpha = 0.05


results = st.ttest_ind(games.query('platform == "XOne" and user_score != 0')['user_score'], \
                       games.query('platform == "PC" and user_score != 0')['user_score'], \
                       equal_var=False)
print('значение p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print('Нулевая гипотеза отвергнута')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Вполне предсказуемый результат - средние пользовательские рейтинги платформ Xbox One и PC отличаются.

### Средние пользовательские рейтинги жанров Action и Sports разные

Нулевой гипотезой будет выражение: 
- "Средние пользовательские рейтинги жанров Action и Sports одинаковые".

Альтернативной гипотезой будет гипотеза: 
- "Средние пользовательские рейтинги жанров Action и Sports различны"

```python
# уровень значимости
alpha = 0.05


results = st.ttest_ind(games.query('genre == "Action" and user_score != 0')['user_score'], \
                       games.query('genre == "Sports" and user_score != 0')['user_score'], \
                       nan_policy='omit')

print('Значение p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print('Нулевая гипотеза отвергнута')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Нам не удалось отпросить нудевую гипотезу и средние пользовательские рейтинги жанров Action и Sports вполне могут быть одинаковыми.
