### Проверка правильности рассчета эффективность обогащения

```python
# скопируем обучающую выборку и избавимся от пропусков
train_data_copy = train_data.copy()
train_data_copy.dropna(inplace = True)


# сохраним необходимые данные в переменных
c = train_data_copy['rougher.output.concentrate_au']
f = train_data_copy['rougher.input.feed_au']
t = train_data_copy['rougher.output.tail_au']

# сохраним исходные данные эффективности обогащения и рассчитаем по имеющимся данным
given_recovery = train_data_copy['rougher.output.recovery']
count_recovery = (c * (f - t) / (f*(c - t))) * 100

# рассчитаем среднюю абсолютную ошибку
mean_absolute_error(given_recovery, count_recovery)
```

Так как средняя ошибка предельно мала - можно сделать вывод, что эффекктивность обогащения в наших данных предоставлена верная.

### Как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки

```python
# запишем функцию для постройки гистограмм распредления для концетрации металлов

def metal_concentration(metal, title):
    x0 = train_data[f'rougher.input.feed_{metal}']
    x1 = train_data[f'rougher.output.concentrate_{metal}']
    x2 = train_data[f'primary_cleaner.output.concentrate_{metal}']
    x3 = train_data[f'final.output.concentrate_{metal}']

    plt.figure(figsize=(14, 8))

    sns.histplot(x0, color='skyblue', alpha=0.5, label='Концентрация в исходном сырье', kde=True)
    sns.histplot(x1, color='darkorchid', alpha=0.5, label='Концентрация после флотации', kde=True)
    sns.histplot(x2, color='tomato', alpha=0.5, label='Концентрация после первичной очистки', kde=True)
    sns.histplot(x3, color='gold', alpha=0.5, label='Финальная концентрация', kde=True)

    plt.title(label=f'Концентрация {title} на разных этапах', fontsize=12)
    plt.xlabel('Концентрация металла')
    plt.ylabel('Количество наблюдений')
    plt.legend()
    plt.show()
```

```python
# посмотрим на концетрация золота

metal_concentration('au', 'золота')
```

```python
# посмотрим на концетрация серебра

metal_concentration('ag', 'серебра')
```

```python
# посмотрим на концетрация свинца 

metal_concentration('pb', 'свинца')
```

- у всех металлов на всех этапах наблюдаются колоссальные выбросы около 0
- концентрация золота (au) поэтапно повышается от первого до последнего этапа 
- концентрация сереребра (ag) на входе больше, чем на выходе 
- концентрация свинца (pb) увеличивается от первого этапа до первичной очистки, но после финального этапа остается прежней

### Исследование суммарной концентраций всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах

```python
# сведем суммарные показетели концентрации по каждому этапу в один столбец

full_data['rougher_input_feed'] = full_data['rougher.input.feed_au'] + \
full_data['rougher.input.feed_ag'] + full_data['rougher.input.feed_pb'] + \
full_data['rougher.input.feed_sol']

full_data['rougher_output_concentrate'] = full_data['rougher.output.concentrate_au'] + \
full_data['rougher.output.concentrate_ag'] + full_data['rougher.output.concentrate_pb'] + \
full_data['rougher.output.concentrate_sol']

full_data['final_output_concentrate'] = full_data['final.output.concentrate_au'] + \
full_data['final.output.concentrate_ag'] + full_data['final.output.concentrate_pb'] + \
full_data['final.output.concentrate_sol']
```

```python
# зададим размер полотна
plt.figure(figsize=(14, 8))

# строим графики
sns.histplot(full_data['rougher_input_feed'], color='red', alpha=0.5, \
             label='Сырье', kde=True)
sns.histplot(full_data['rougher_output_concentrate'], color='lime', alpha=0.5, \
             label='Черновой концентрат', kde=True)
sns.histplot(full_data['final_output_concentrate'], color='fuchsia', alpha=0.5, \
             label='Финальный концентрат', kde=True)



# задаем названия и размеры шрифта
plt.title('Суммарная концентрация веществ на различных этапах', fontsize=12)
plt.xlabel('Концентрация веществ')
plt.ylabel('Количество наблюдений')
plt.legend()
plt.show()
```

Как и случае с частными случаями концентрации металлов на разных этапах - суммарная концентрация дает нам большие выбросы около нуля. Придется с ними разобраться.

И вот если обратиться к нашему проекту:<br>
1) Есть нулевые аномалии в суммарных концентрациях - и их надо "убить". Дело в том, что предполагается, что хоть грамм чего-нибудь но есть в руде. А если мы имеем ноль, значит большая вероятность неточности измерений.<br>
    
2)А вот с выбросами как в приниципе..<br>
Тут на обсуждение: наша цель - предсказывать модель на реальных данных. И вот эти выбросы - это и есть реальная концентрация. И нам важно, чтобы модель умела и их тоже видеть и обрабатывать. Если их убрать из модели, то получится, что у нас она станет как бы искусственной. Т.е. будет работать только в небольших диапазонах - потому что обучалась только в них.<br>
    
Поэтому, тут так: выбросы - это не всегда то, от чего надо избавляться))) Ещё зависит от бизнес-задачи.<br>
    
Кстати<br>
если говорить о теории выбросов с помощью графика "ящик с усами", то вот это интервал q3+-1.5*iqr говорит действильно о них самых<br>
Но нет рекоммендаций к их удалению. Говорят так "на усмотрение бизнеса".<br>
А вот если взять вот этот интервал q3+3*iqr:  то тут уж с очень высокой вероятность можно говорить о грубых выбросах.

## Модель

### Функции для вычисления sMAPE и  итоговой sMAPE

```python
def smape(target, predict):
    return 1/len(target)*np.sum(2*np.abs(target - predict) / (np.abs(target) + np.abs(predict))*100)
```

```python
def final_smape(rougher, final):
    return 0.25*rougher + 0.75*final
```

```python
smape_score = make_scorer(smape, greater_is_better=True)
```

#### Расчет итогового sMAPE

```python
# считаем sMAPE для флотации и финальных данных

y_test_rougher = full_data[full_data.index.isin(test_data_clean.index)]['rougher.output.recovery']
y_test_final = full_data[full_data.index.isin(test_data_clean.index)]['final.output.recovery']

lr_smape_rougher = smape(y_test_rougher, lr_predict_rougher)
forest_smape_final = smape(y_test_final, forest_predict_final)
```

```python
# считаем итоговый sMAPE

final_smape(lr_smape_rougher, forest_smape_final)
```

```python
# проверим наш результат на константной dummy-модели 

dummy = DummyRegressor(strategy="mean")
dummy_rougher = dummy.fit(x_train, y_train_rougher)
dummy_final = dummy.fit(x_train, y_train_final)

dummy_rougher_smape = smape(y_test_rougher, dummy_rougher.predict(x_test))
dummy_final_smape = smape(y_test_final, dummy_final.predict(x_test))

final_smape(dummy_rougher_smape, dummy_final_smape)
```

Итоговый sMAPE наших вычислений на двух моделях оказался чуть лучше результата константной модели.
